```python
import pandas as pd
import os
import re
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
os.listdir("data")
```

```python
# load categories data into csv

output_file = open('data/categories.txt', 'a')
output_file.writelines("page, category \n")

filepath = 'data/article_categories_en.nt'
with open(filepath) as fp:
    line = fp.readline()
    cnt = 1
    while line:
#         print("Line {}: {}".format(cnt, line.strip()))
        line = fp.readline()
        cnt += 1
        
        if cnt % 100000 == 0:
            print(cnt)
        
        # skip header
        if cnt > 0:
            try:
                key = line.split(" ")[0].split(">")[0].split("<http://dbpedia.org/resource/")[1]
                value = line.split(" ")[2].split(">")[0].split("<http://dbpedia.org/resource/")[1].split("Category:")[1]
                
                key = re.sub(",", "__", key)
                value = re.sub(",", "__", value)
                
                output_file.writelines(key + "," + value + "\n")
            except:
                pass

output_file.close()
```
